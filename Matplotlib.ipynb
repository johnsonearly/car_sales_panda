```python
import numpy as np
x = np.linspace(0,10,100)
x[:15]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2)
ax.set(title = "Line plot",
      xlabel = "x",
      ylabel = "x^2")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
ax.set(title = "Scatter plot")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
plt.show()
```

```python
nut_butter_prices = {"Almond":10,
                   "Peanut butter": 7,
                   "Cashew butter": 4}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),height = nut_butter_prices.values())
ax.set(title = "Johnson's butter sales",
      ylabel = "Prices",)
plt.show()
```

```python
fig,ax = plt.subplots()
#We cant use barh on dictionaries since they are unhashable
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
plt.show()
```

```python
#Make some data for histogram
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x)
plt.show()
```

<mark><h1> SUBPLOTS</h1></mark>
<p>There are two options for subplots</p>

```python
#Option1
np.random.seed(0)
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,ncols =2,figsize = (10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(20),np.random.random(20))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(x)
plt.show()
```

```python
#Option 2
np.random.seed(2)
fig,ax = plt.subplots(nrows = 2,ncols =2,figsize = (10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(20),np.random.random(20))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(x)
plt.show()
```

## Which method to use
The matplotlib 00  method is preferrable for complex data visualization but for simpler cases we can use the pyplot method

```python
#How many are over 50
len(over_50)
```

```python
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
            cmap = "winter")
```

```python
#00 Method
fig,ax = plt.subplots(figsize = (10,5))
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax = ax);
```

```python
fig,ax = plt.subplots(figsize = (10,5))
scatter = ax.scatter( x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])
ax.set(title = "Heart Disease",
       xlabel = "age(yrs)",
       ylabel = "cholestrol()")
ax.legend(*scatter.legend_elements(),title = "Target")
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
fig,(ax0,ax1)= plt.subplots(nrows =2,
                            ncols = 1,
                            figsize = (10,10),
                            sharex = True)
scatter = ax0.scatter( x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"] )
ax0.set(title = "Heart Disease and Cholestrol Level",
       ylabel = "cholestrol()")
ax0.legend(*scatter.legend_elements(),title = "Target")
ax0.axhline(over_50["chol"].mean(),
           linestyle = "--");

scatter = ax1.scatter( x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
ax1.set(title = "Heart Disease and Max Heart Rate",
       xlabel = "age(yrs)",
       ylabel = "thalach()")
ax1.legend(*scatter.legend_elements(),title = "Target")
ax1.axhline(over_50["thalach"].mean(),
           linestyle = "--");
fig.suptitle("HEART DISEASE ANALYSIS ", fontweight = "bold") ;
```
